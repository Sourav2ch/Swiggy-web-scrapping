```python
restro2=restro10.set_index(keys='Cuisine')
restro2
```

```python
cuisine1=restro2.index # get cuisine values from restro
cuisine1
```

```python
#Program to get only the unique values in Cuisine
cuisine2=[]
for i in cuisine1:
    cuisine2.append(i)
c1=[]
for i in cuisine2:
    a=re.findall(r'(.+?),',i)
    for j in a:
        c1.append(j)  
```

```python
print(c1)
```

```python
c2=set(c1)
c3=list(c2)
```

```python
print(c3)
len(c3)
```

```python
c4=[]
for i in c3:
    c4.append(i.strip())
```

```python
c5=set(c4)
c6=list(c5)
len(c5)
c6.pop(0)
print(c6)
```

```python
s1=[] ### Number of hotels available for that specified cuisine
for i in c6:
    restro4=restro2.filter(like=i,axis=0)
    restro4.reset_index(inplace=True)
    restro4=restro4.drop_duplicates('Hotel')
    s1.append(restro4['Hotel'].count())
print(s1)
```

```python
Dish=pd.DataFrame()
Dish['Cuisine']=c6
Dish['Count']=s1
Dish.head(40)
```

```python
plt.figure(figsize=(10,100))
sns.catplot(x="Cuisine",y="Count",data=Dish,kind="bar",height=4,aspect=5)
plt.xticks(rotation=1000)
```
