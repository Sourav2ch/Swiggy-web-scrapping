```python
#Program to get only the unique values in Cuisine
cuisine2=[]
for i in cuisine1:
    cuisine2.append(i)
c1=[]
for i in cuisine2:
    a=re.findall(r'(.+?),',i)
    for j in a:
        c1.append(j)  
```
