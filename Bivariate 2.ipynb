```python
#Program to get only the unique values in Cuisine
cuisine2=[]
for i in cuisine1:
    cuisine2.append(i)
c1=[]
for i in cuisine2:
    a=re.findall(r'(.+?),',i)
    for j in a:
        c1.append(j)  
```

>>> - From the above joint plot it is observed that the data collected is NORMAL DISTRIBUTION

> - From the uni-variate analysis done before it is observed that top cuisines 
are NorthIndian , South Indian , Chinese , Fast Food , Biryani.
> - Lets observe the average cost and rating for these Cuisines. 

```python
j=0
restro4=restro2.filter(like='North Indian',axis=0)
restro4.reset_index(inplace=True)

for i in restro4['Cuisine']:
    restro4.loc[j,'Cuisine']='North Indian'
    j+=1
k=0
restro5=restro2.filter(like='South Indian',axis=0)
restro5.reset_index(inplace=True)

for i in restro5['Cuisine']:
    restro5.loc[k,'Cuisine']='South Indian'
    k+=1
l=0
restro6=restro2.filter(like='Fast Food',axis=0)
restro6.reset_index(inplace=True)

for i in restro6['Cuisine']:
    restro6.loc[l,'Cuisine']='Fast Food'
    l+=1
m=0
restro7=restro2.filter(like='Biryani',axis=0)
restro7.reset_index(inplace=True)

for i in restro7['Cuisine']:
    restro7.loc[m,'Cuisine']='Biryani'
    m+=1
n=0
restro8=restro2.filter(like='Chinese',axis=0)
restro8.reset_index(inplace=True)

for i in restro8['Cuisine']:
    restro8.loc[n,'Cuisine']='Chinese'
    n+=1
```

```python
restro3=pd.concat([restro4,restro5,restro6,restro7,restro8])
restro3
restro3['Cuisine'].unique()
```
