### Nizampet

```python



for nums in range(0,7):
    data = requests.get('https://www.swiggy.com/hyderabad/nizampet-pragathi-nagar-restaurants?page={}'.format(nums))
    
    data_soup=BeautifulSoup(data.text)
    restaurant_tag=data_soup.find_all('div',attrs={"class": "_3FR5S"})
    
    
    for i in restaurant_tag:
        
        hotel=i.find('div',class_='nA6kb').text
        if hotel:
            Hotel.append(hotel)
            print(hotel)
        
        cuisine=i.find('div',class_='_1gURR').text
        if cuisine:
            Cuisine.append(cuisine)
            print(cuisine)
        else:
            print("null")
        
        rating=i.find('div',class_=['_9uwBC _2lAlc','_9uwBC wY0my','_9uwBC']).text
        if rating:
            Rating.append(rating)
            print(rating)
        
        cost=i.find('div',class_=["nVWSi"]).text
        if cost:
            Cost.append(cost)
            print(cost)
        
        area_tag=data_soup.find('span',attrs={"class": "kpkwa"}).text
        if area_tag:
            Area.append(area_tag)
            print(area_tag)

    
```
